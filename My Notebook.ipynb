```python
import matplotlib.pyplot as plt
```

```python
heart_rates = [72, 75, 78, 80, 77, 74, 79, 81, 83, 76]  # Example data
indices = list(range(1, len(heart_rates) + 1))  # 1-based index for better readability
```

```python
len(heart_rates)
```

```python
list(range(1, 10))
```

```python
indices
```

```python
fig, ax = plt.subplots()
plt.plot(indices, heart_rates, marker='o', linestyle='-', color='red')
plt.xlabel("Index")
plt.ylabel("Heart Rate (BPM)")
plt.title("Heart Rate Readings")
plt.grid(True)
plt.show()
fig.savefig("images/heart_rate_line_plot.png")
```

```python
fig, ax = plt.subplots()
plt.hist(heart_rates, bins=5, color='blue', edgecolor='black', alpha=0.7)
plt.xlabel("Heart Rate (BPM)")
plt.ylabel("Frequency")
plt.title("Heart Rate Distribution")
plt.show()
fig.savefig("images/heart_rate_histogram.png")
```

```python
fig, ax = plt.subplots()
plt.boxplot(heart_rates, vert=True, patch_artist=True)
plt.ylabel("Heart Rate (BPM)")
plt.title("Heart Rate Variability")
plt.show()
fig.savefig("images/heart_rate_box_plot.png")
```

```python
fig, ax = plt.subplots()
plt.scatter(indices, heart_rates, color='red', marker='o')
plt.xlabel("Index")
plt.ylabel("Heart Rate (BPM)")
plt.title("Heart Rate Scatter Plot")
plt.grid(True)
plt.show()
fig.savefig("images/heart_rate_scatter_plot.png")
```
